#**Step 6: Filling Missing Values**

### **Step: This process is part of data cleaning under EDA (Exploratory Data Analysis). Handling null values makes your dataset more consistent and avoids errors during visualization or further analysis.**

# Handling Missing (Null) Values in a Product Column

```python
df[df['product'].isnull()]
```

```python
# The .mode() function returns the most frequently occurring value(s) in the 'product' column.
df['product'].mode()
```

```python
# This command replaces all null (missing) values in the 'product' column with the string "Turmeric Powder/Arisina Pudi".
df['product']=df['product'].fillna('Turmeric Powder/Arisina Pudi')
```

```python
df['product'].isnull().sum()
```

# Step: Handling Missing (Null) Values in a Brand Column

```python
df[df['brand'].isnull()]
```

```python
# The .mode() function returns the most frequently occurring value(s) in the 'product' column.
df['brand'].mode()
```

```python
# This command replaces all null (missing) values in the 'Brand' column with the string "Fresha".
df['brand']=df['brand'].fillna('Fresha')
```

# Handling Missing (Null) Values in a Sale Price Column.

```python
df['sale_price'].dtypes
```

```python
# This command calculates the median value of the 'sale_price' column.
df['sale_price'].median()
```

```python
# Filling Missing Values in the 'sale_price' Column.
df['sale_price']=df['sale_price'].fillna(190.0)
```

# Handling Missing (Null) Values in a Rating Column.

```python
df['rating'].isnull().sum()
```

```python
df.duplicated().sum()
```

```python
#  This command calculates the median value of the 'Rating' column.
df['rating'].median()
```

```python
# # Filling Missing Values in the 'Rating' Column.
df['rating'] = df['rating'].fillna(df['rating'].median())
```

# Dropping Rows with Missing Values in the 'description' Column.

```python
# Counting Missing Values in the 'description' Column
df['description'].isnull().sum()
```

```python
# This command removes all rows from the DataFrame where the 'description' column has null (missing) values
df= df.dropna(subset=['description'])
```

# Outliers Detection and Fill with Mean

```python
# Example: Handling outliers in 'sale_price' column
mean_price = df['sale_price'].mean()
```

```python
mean_price
```

# This line limits the values in the sale_price column of the DataFrame df to a maximum (upper) threshold.



```python
# If any value in sale_price is greater than 3 * mean_price, it is replaced with 3 * mean_price.
# All other values remain unchanged
df['sale_price'] = df['sale_price'].apply(lambda x: mean_price if x > 3*mean_price else x)
```

```python
# maximum value in the sale_price column.
df['sale_price'].max()
```

```python
df[df['sale_price'] == df['sale_price'].max()]
```

```python
# maximum value in the sale_price column
df['sale_price'].min()
```

```python
df[df['sale_price'] == df['sale_price'].min()]
```

# **Step 10 : Simple Plots or Visualizations**

```python
# The box represents the interquartile range (middle 50% of the data).
# The line inside the box is the median sale price.
# Whiskers extend to show the rest of the distribution (excluding outliers).
# Dots beyond the whiskers are outliers  unusually high or low sale prices.

plt.figure(figsize=(6, 4))
sns.boxplot(x=df['sale_price'])
plt.title('Boxplot of Sale Price')
plt.show()
```

#  Sale Price Distribution (Histogram with **KDE**)

```python
# This histogram shows the distribution of sale prices.

plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
sns.histplot(data=df, x='sale_price', bins=40, palette='Set1', kde=True,color='skyblue')
plt.title("Sale Price Histogram")
plt.xlabel("Sale Price Ranges (divided into 40 bins)")
plt.ylabel("Number of Products (Frequency)")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

# This **correlation heatmap** visually represents the relationships between the numerical features in the dataset.


1.   The heatmap is based on Pearson correlation coefficients, which **range from -1 to +1.**


*   **+1** *indicates a perfect positive linear relationship.*
*   **-1** *indicates a perfect negative linear relationship.*
*   ** 0** *means no linear correlation*.


  



```python
corr = numeric_df.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(corr, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```
